PANDAS

Pandas is a Python library used for data analysis and manipulation. It provides powerful data structures like DataFrame and Series to work with tabular data efficiently.

```python
# Import Pandas Library

import pandas as pd 

# Read CSV File

df = pd.read_csv("titanic.csv")

print("Dataset Loaded Successfully")
```

1. head()

The head() method displays the first 5 rows of the dataset by default.

```python
print(df.head())
```

2. tail()

The tail() method displays the last 5 rows of the dataset by default.

```python
print(df.tail())
```

3. shape

The shape attribute returns the number of rows and columns in the dataset.

```python
print("Shape of Dataset:", df.shape)
```

4. columns

The columns attribute returns the names of all columns in the dataset.

```python
print("Columns:", df.columns)
```

5. info()

The info() method provides information about the dataset such as data types, non-null values, and memory usage.

```python
df.info()
```

6. describe()

The describe() method gives a statistical summary of numerical columns.

```python
print(df.describe())
```

7. isnull().sum()

The isnull().sum() method counts the missing values in each column.

```python
print(df.isnull().sum())
```

8. value_counts()

The value_counts() method counts the occurrence of unique values in a column.

```python
print(df["Sex"].value_counts())
```

9. sort_values()

The sort_values() method sorts the dataset based on a specified column.

```python
print(df.sort_values(by="Age"))
```

10. sample()

The sample() method returns random rows from the dataset.

```python
print(df.sample(5))
```
