Green Skill / Sustainability Dataset Analysis

A sustainability dataset contains information related to environmental factors such as carbon emissions, renewable energy, air quality, population, and other green indicators. It helps us analyze environmental trends and make informed decisions.

```python
# Import Pandas Library

import pandas as pd

df = pd.read_csv("owid-co2-data.csv")

print("Dataset Loaded Successfully")
```

1. head()

The head() method displays the first 5 rows of the dataset.

```python
print(df.head())

# Analysis:
# The first five rows show the initial records of the dataset.
# Each row represents a country's data for a particular year.
```

2. tail()

The tail() method displays the last 5 rows of the dataset.

```python
print(df.tail())

# Analysis:
# The last five rows display the ending records of the dataset.
```

3. shape

The shape attribute returns the number of rows and columns in the dataset.

```python
print("Shape of Dataset:", df.shape)

# Analysis:
# The dataset contains many rows and columns related to sustainability.
```

4. columns

The columns attribute returns the names of all columns in the dataset.

```python
print(df.columns)

# Analysis:
# Every column represents a different environmental feature such as
# country, year, CO2 emissions, population, GDP, etc.
```

5. info()

The info() method displays information about the dataset.

```python
df.info()

# Analysis:
# It shows the data type of each column,
# number of non-null values,
# and memory usage.
```

6. describe()

The describe() method gives statistical information about numerical columns.

```python
print(df.describe())

# Analysis:
# It displays count, mean, minimum,
# maximum and other statistical values.
```

7. isnull().sum()

The isnull().sum() method counts the missing values in every column.

```python
print(df.isnull().sum())

# Analysis:
# Some columns contain missing values
# because data may not be available for every country and year.
```

8. Handle Null Values

Missing values can affect data analysis. Here, numerical null values are replaced with the mean of their respective columns.

```python
numeric_columns = df.select_dtypes(include="number").columns

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

print(df.isnull().sum())

# Analysis:
# Numerical missing values are replaced with the mean.
# This helps in performing analysis without losing data.
```

9. sample()

The sample() method displays random rows from the dataset.

```python
print(df.sample(5))

# Analysis:
# Random records help us inspect different
# parts of the dataset.
```

10. Generalisation and Relationship Between Columns

Different columns in the dataset are related to each other and help in understanding sustainability trends.

```python
# Generalisation

# Country:
# Represents the name of the country.

# Year:
# Represents the year in which the data was recorded.

# CO2:
# Represents carbon dioxide emissions.

# Population:
# Countries with higher populations may have higher CO2 emissions.

# GDP:
# Economic growth generally increases energy consumption,
# which can increase emissions.

# Renewable Energy:
# Greater use of renewable energy may reduce carbon emissions.

# Energy Consumption:
# Higher energy consumption generally leads to higher CO2 emissions.

```
